```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("titanic")
df.head()
```

```python
df.select_dtypes(include="object").columns  
```

```python
df.select_dtypes(include=["int64", "float64"]).columns
```

```python
sns.countplot(x="sex", data=df)
plt.title("Bar Chart of Sex")
plt.show()
```

```python
df["class"].value_counts().plot(
    kind="pie",
    autopct="%1.2f%%",
    startangle=90
)
plt.title("Pie Chart of Passenger Class")
plt.ylabel("")
plt.show()
```

```python
plt.scatter(df.index, df["age"])
plt.title("Scatter Plot of Age")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

```python
plt.plot(df["fare"])
plt.title("Line Plot of Fare")
plt.xlabel("Index")
plt.ylabel("Fare")
plt.show()
```

```python
sns.stripplot(x=df["pclass"], y=df["age"])
plt.title("Strip Plot of Pclass and Age")
plt.show()
```
